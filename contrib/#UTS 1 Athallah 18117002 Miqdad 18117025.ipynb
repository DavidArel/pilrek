```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import re
import plotly.express as px
```

```python
url = 'https://github.com/eueung/pilrek/raw/master/pilrek.csv'
df  = pd.read_csv(url)
```

# Visualisasi 1
## Berikut merupakan TOP 5 calon pilihan rektor berdasarkan voting kapabilitasnya masing-masing

```python
a = df.loc[df['Alasan Memilih CaRek'].str.contains('^kapabilitas[a-z]*', flags=re.I, regex=True)]
a['CaRek Pilihan'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%')
```

# Visualisasi 2
## Perbandingan kategori pemilih terhadap calon pilihan rektor random

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
a1 = df.loc[df['CaRek Pilihan'] == "Acep Purqon (FMIPA-FI)"]
a1['Kategori Anda'].value_counts().plot(kind='barh', color = 'chartreuse', ax=axes[0,0]).set_title('Acep Purqon')
a2 = df.loc[df['CaRek Pilihan'] == "Jaka Sembiring (STEI)"]
a2['Kategori Anda'].value_counts().plot(kind='barh', color = 'chartreuse', ax=axes[0,1]).set_title('Jaka Sembiring')
a3 = df.loc[df['CaRek Pilihan'] == "Edy Tri Baskoro (FMIPA-MA)"]
a3['Kategori Anda'].value_counts().plot(kind='barh', color = 'chartreuse', ax=axes[1,0]).set_title('Edy Tri Baskoro')
a4 = df.loc[df['CaRek Pilihan'] == "Ary Setiaji Prihatmanto (STEI)"]
a4['Kategori Anda'].value_counts().plot(kind='barh', color = 'chartreuse', ax=axes[1,1]).set_title('Ary Setiaji Prihatmanto')
fig.tight_layout()
plt.show()
```

# Visualisasi 3 
## Top 5 calon pilihan rektor terhadap jumlah pemilihnya yang berkategori umur < 25

```python
a = df.loc[df['Kategori Umur Anda'] == "< 25 th"]
b = a['CaRek Pilihan'].value_counts().head(5)
c = pd.DataFrame(dict(
   r=[66, 28, 23, 8, 8],
    theta=['Acep Purqon (FMIPA-FI) ','Edy Tri Baskoro (FMIPA-MA)','I Nyoman Pugeg Aryantha (SITH)', 
           'Khairurrijal (FMIPA-FI)', 'Ary Setiaji Prihatmanto (STEI)']))
fig = px.line_polar(c, r = 'r', theta ='theta', line_close = True)
fig.update_traces(fill='toself')
fig.show()
```
