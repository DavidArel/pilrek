```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('10_Bakal_Calon_Rektor.csv')
data_2 = pd.read_csv('24_Calon_Rektor_Tersingkir.csv')
```

```python
persen_kategori = data['Kategori Anda'].value_counts()
```

```python
persen_kategori.plot(kind= 'bar', x = 'Kategori Anda')
```

```python
persen_kategori.plot(kind= 'pie', x = 'Kategori Anda', autopct='%1.1f%%')
```

```python
kategori = data['Kategori Anda'].value_counts()
kategori_2 = data_2['Kategori Anda'].value_counts()

sum = 0
for i in kategori:
    sum += i

sum_2 = 0
for i in kategori_2:
    sum_2 += i

persen_kategori_list = []
for i in range(len(kategori)):
    persen_kategori_list.append(kategori[i]/sum)

persen_kategori_list_2 = []
for i in range(len(kategori_2)):
    persen_kategori_list_2.append(kategori_2[i]/sum_2)

persen_kategori_series = pd.Series(persen_kategori_list, index = kategori.index)
persen_kategori_series_2 = pd.Series(persen_kategori_list_2, index = kategori_2.index)

persen_kategori_dataframe = pd.concat([persen_kategori_series, persen_kategori_series_2],sort=True, axis = 1, keys=['Rektor_terpilih', 'Rektor_tidak_terpilih'])

persen_kategori_dataframe = persen_kategori_dataframe.sort_values(by='Rektor_terpilih', ascending=False)
```

```python
persen_kategori_dataframe
```

```python
persen_kategori_dataframe.plot(kind='bar', y=['Rektor_terpilih', 'Rektor_tidak_terpilih'])
```

```python
umur = data['Kategori Umur Anda'].value_counts()
umur_2 = data_2['Kategori Umur Anda'].value_counts()

sum = 0
for i in umur:
    sum += i

sum_2 = 0
for i in umur_2:
    sum_2 += i

persen_umur_list = []
for i in range(len(umur)):
    persen_umur_list.append(umur[i]/sum)

persen_umur_list_2 = []
for i in range(len(umur_2)):
    persen_umur_list_2.append(umur_2[i]/sum_2)

persen_umur_series = pd.Series(persen_umur_list, index = umur.index)
persen_umur_series_2 = pd.Series(persen_umur_list_2, index = umur_2.index)

persen_umur_dataframe = pd.concat([persen_umur_series, persen_umur_series_2],sort=True, axis = 1, keys=['Rektor_terpilih', 'Rektor_tidak_terpilih'])

persen_umur_dataframe = persen_umur_dataframe.sort_values(by='Rektor_terpilih', ascending=False)    
```

```python
alasan_value = data['Alasan Memilih CaRek'].value_counts()
alasan_value_2 = data_2['Alasan Memilih CaRek'].value_counts()

alasan_with_other = []
for alasan in alasan_value.index[:12]:
    alasan_with_other.append(alasan)
alasan_with_other.append('Other')

alasan_with_other_2 = []
for alasan_2 in alasan_value_2.index[:12]:
    alasan_with_other_2.append(alasan_2)
alasan_with_other_2.append('Other')

jumlah_alasan = []
for i in alasan_value[:12]:
    jumlah_alasan.append(i)
count_other = 0
for i in alasan_value[12:]:
    count_other += i
jumlah_alasan.append(count_other)

jumlah_alasan_2 = []
for i in alasan_value_2[:12]:
    jumlah_alasan_2.append(i)
count_other_2 = 0
for i in alasan_value_2[12:]:
    count_other_2 += i
jumlah_alasan_2.append(count_other_2)

sum = 0
for i in jumlah_alasan:
    sum += i
for i in range(len(jumlah_alasan)):
    jumlah_alasan[i] = jumlah_alasan[i]/sum

sum_2 = 0
for i in jumlah_alasan_2:
    sum_2 += i
for i in range(len(jumlah_alasan_2)):
    jumlah_alasan_2[i] = jumlah_alasan_2[i]/sum_2

alasan_other_series = pd.Series(jumlah_alasan, index=alasan_with_other)
alasan_other_series_2 = pd.Series(jumlah_alasan_2, index=alasan_with_other_2)

alasan_gabung = pd.concat([alasan_other_series, alasan_other_series_2],sort=True, axis = 1, keys=['Rektor_terpilih', 'Rektor_tidak_terpilih'])

alasan_gabung = alasan_gabung.sort_values(by='Rektor_terpilih', ascending=False)
```

```python
alasan_gabung.plot(kind='bar', y=['Rektor_terpilih', 'Rektor_tidak_terpilih'])
```
