```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url1 = 'https://github.com/eueung/pilrek/raw/master/pilrek.csv'
df1  = pd.read_csv(url1)
df1.head()
```

```python
url2 = 'https://github.com/eueung/pilrek/raw/master/pilrek-anon.csv'
df2  = pd.read_csv(url2)
df2.head()
```

## Menggabungkan data anon dan data non-anon

```python
df = pd.concat([df1,df2])
```

```python
# data frame untuk pengisi <25 thn
df25 = df[df['Kategori Umur Anda'] == '< 25 th']
```

```python
# data frame untuk pengisi 25-40 thn
df2540 = df[df['Kategori Umur Anda'] == '25 - 40 th']
```

```python
# data frame untuk pengisi >40 thn
df40 = df[df['Kategori Umur Anda'] == '> 40 th']
```

# Visualisasi Umum

# Profil 10 finalis rektor di mata para responden

## Data ini murni diambil dari data responden

```python
calon1 = df[(df['CaRek Pilihan'] == 'Jaka Sembiring (STEI)')]
calon2 = df[df['CaRek Pilihan'] == 'Benyamin Sapiie (FITB)']
calon3 = df[df['CaRek Pilihan'] == 'Bramantyo (Direktur PPM Manajemen)']
calon4 = df[df['CaRek Pilihan'] == 'Dwi Larso (SBM / President Univ)']
calon5 = df[df['CaRek Pilihan'] == 'Edy Tri Baskoro (FMIPA-MA)']
calon6 = df[df['CaRek Pilihan'] == 'Gusti Ayu Putri Saptawati S (STEI)']
calon7 = df[df['CaRek Pilihan'] == 'Kadarsah Suryadi (FTI-TI)']
calon8 = df[df['CaRek Pilihan'] == 'Reini D Wirahadikusumah (FTSL)']
calon9 = df[df['CaRek Pilihan'] == 'Togar M Simatupang (SBM)']
calon10 = df[df['CaRek Pilihan'] == 'Widjaja Martokusumo (SAPPK)']
```

# PENGOLAHAN UNTUK CALON 1

```python
kri1 = calon1[(calon1['Alasan Memilih CaRek'] == 'Keberhasilan dan prestasi')]
kri2 = calon1[(calon1['Alasan Memilih CaRek'] == 'Sesuai dengan kriteria kekinian (Rektor 4.0)')]
kri3 = calon1[(calon1['Alasan Memilih CaRek'] == 'Kecerdasan dan keberanian utk kemajuan ITB')]
kri4 = calon1[(calon1['Alasan Memilih CaRek'] == 'Kapabilitas ybs utk memajukan riset dan industri RI')]
kri5 = calon1[(calon1['Alasan Memilih CaRek'] == 'Karakter kepemimpinan dan leadership')]

labels=np.array(['Prestasi', 'Rektor 4.0', 'Kecerdasan', 'Kapabilitas memajukan Riset', 'Kepemimpinan'])
stats = [kri1.count(), kri2.count(), kri3.count(), kri4.count(), kri5.count()]

angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
stats=np.concatenate((stats,[stats[0]]))
angles=np.concatenate((angles,[angles[0]]))

fig=plt.figure()
ax = fig.add_subplot(polar=True)
ax.plot(angles, stats, 'o-', linewidth=2)
ax.set_rlabel_position(90)
ax.fill(angles, stats, alpha=0.25)
ax.set_thetagrids(angles * 180/np.pi, labels)
ax.set_title('Jaka Sembiring')
ax.grid(True)
```
