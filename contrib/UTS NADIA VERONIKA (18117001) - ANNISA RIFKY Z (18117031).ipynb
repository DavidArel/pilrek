```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://github.com/eueung/pilrek/raw/master/pilrek-anon.csv'
df  = pd.read_csv(url)
```

```python
new = df.groupby('Kategori Anda').get_group('Dosen')
new["Alasan Memilih CaRek"].value_counts().plot(kind='bar')
plt.xlabel('Alasan Memilih')
plt.ylabel('Jumlah Pemilih')
plt.title('Alasan Para Dosen Dalam Memilih Carek')
```

```python
word = input ("Pilih Kategori")
if word == 'Dosen':
    new = df.groupby('Kategori Anda').get_group('Dosen')
    new["CaRek Pilihan"].value_counts().head(5).plot(kind='pie',autopct='%1.1f%%')
elif word == 'Pegawai / Tendik':
    new = df.groupby('Kategori Anda').get_group('Pegawai / Tendik')
    new["CaRek Pilihan"].value_counts().head(5).plot(kind='pie',autopct='%1.1f%%')
elif word == 'Mahasiswa':
    new = df.groupby('Kategori Anda').get_group('Mahasiswa')
    new["CaRek Pilihan"].value_counts().head(5).plot(kind='pie',autopct='%1.1f%%')
elif word == 'Alumni':
    new = df.groupby('Kategori Anda').get_group('Alumni')
    new["CaRek Pilihan"].value_counts().head(5).plot(kind='pie',autopct='%1.1f%%')
elif word == 'Umum':
    new = df.groupby('Kategori Anda').get_group('Umum')
    new["CaRek Pilihan"].value_counts().head(5).plot(kind='pie',autopct='%1.1f%%')
  
```

```python
print ("Untuk menampilkan 5 CaRek Terpopuler di Kalangan Umur 25-40 th, ketik : Y")
print ("Untuk menampilkan 5 CaRek Paling Tidak Populer di Kalangan Umur 25-40 th, ketik : N")
word = input ("Pilihan :")
if word == 'Y':
    new_1 = df.groupby('Kategori Umur Anda').get_group('25 - 40 th')
    new_1["CaRek Pilihan"].value_counts().head(5).plot(kind='bar')
    plt.xlabel('CaRek Pilihan')
    plt.ylabel('Jumlah Pemilih')
    plt.title('5 CaRek Paling Populer di Kalangan Umur 25-40 th')
elif word == 'N':
    new_2 = df.groupby('Kategori Umur Anda').get_group('25 - 40 th')
    new_2["CaRek Pilihan"].value_counts().tail(5).plot(kind='bar')
    plt.xlabel('CaRek Pilihan')
    plt.ylabel('Jumlah Pemilih')
    plt.title('5 CaRek Paling Tidak Populer di Kalangan Umur 25-40 th')
```
