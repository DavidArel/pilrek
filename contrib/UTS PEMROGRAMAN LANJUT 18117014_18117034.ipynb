```python
data_acep = data.groupby('CaRek Pilihan').get_group('Acep Purqon (FMIPA-FI)')
data_acep
```

```python
data_acep.sort_values(by='Alasan Memilih CaRek')
```

```python
data_acep.groupby('Alasan Memilih CaRek').count()
```

```python
acep_skill = data_acep.groupby('Alasan Memilih CaRek').count()[['CaRek Pilihan']].transpose()
acep_skill
```

```python
import numpy as np
```

```python
labels=np.array(['Keberhasilan dan prestasi', 'Mendatangkan dana riset dan kerjasama eksternal, dalam dan luar negeri', 'Muda dan berenergi', 'Sesuai dengan kriteria kekinian (Rektor 4.0)', 'Tipe yg futuristik dan outside-the-box yg bisa menangkap peluang masa depan'])
stats= acep_skill.loc['CaRek Pilihan', labels].values
stats
```

```python
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
stats=np.concatenate((stats,[stats[0]]))
angles=np.concatenate((angles,[angles[0]]))
```

```python
import seaborn as sns
```

```python
fig=plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, stats, 'o-', linewidth=2)
ax.fill(angles, stats, alpha=0.25)
ax.set_thetagrids(angles * 180/np.pi, labels)
ax.set_title('Acep Furqon Top 5 Skill')
ax.grid(True)
```
