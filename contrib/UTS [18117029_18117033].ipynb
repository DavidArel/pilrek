```python
import pandas as pd #Import library Pandas
import matplotlib.pyplot as plt #Import library Matplotlib
%matplotlib inline
url = 'https://github.com/eueung/pilrek/raw/master/pilrek.csv'
df  = pd.read_csv(url, delimiter = ',', names = ['Timestamp','Kategori','Umur','Pilihan','Alasan','Harapan']) #Memberi header
```

```python
# Ambil Data yang mengisi Harapan diKosongkan
kosong = df[['Kategori','Umur','Pilihan','Harapan']].fillna('-')
kosong = kosong[kosong.Harapan == '-'].reset_index(drop = True)
kosong
```

```python
# Visualisasi 1
# Memvisualisasikan banyaknya suara harapan kosong utk tiap calon rektor
vis1 = kosong.groupby('Pilihan').count().sort_values(by = ['Harapan'],ascending = False)
vis1 = vis1[['Harapan']]
vis1
```

```python
vis1.plot(kind='bar',    # Plot a bar chart
        width=0.75,      # Mengatur lebar bar menjadi 75% dari space yang tersedia
        figsize=(10,6),  # Mengatur ukuran plot dalam inchi
        colormap='Accent', #Mengganti warna bar menjadi Hijau Muda
        title = 'Jumlah Harapan Kosong yang Didapat Masing-Masing Ca-Rek') #Memberi judul
```

```python
# Visualisasi Kedua
# Memvisualisasikan 5 Ca-Rek yang paling banyak mendapatkan Harapan Kosong
vis2 = vis1.head()
vis2
```

```python
vis2.plot(kind='barh',    # Plot a bar chart
        width=0.75,      # Mengatur lebar bar menjadi 75% dari space yang tersedia
        figsize=(6,4),  # Mengatur ukuran plot dalam inchi
        colormap='PuOr', #Mengganti warna bar menjadi Coklat
        title = '5 Ca-Rek yang paling banyak mendapatkan Harapan Kosong') #Memberi judul
```

```python
# Visualisasi Ketiga
# Memvisualisasikan Jumlah Harapan Kosong 5 Top Calon Rektor
# Top Calon Rektor adalah Ca-Rek yang paling banyak menerima vote

suara = df.groupby('Pilihan').count().sort_values(by = ['Timestamp'],ascending = False).head()
suara = suara[['Timestamp']] #Variabel 5 calon rektor dgn suara terbanyak
suara
```

```python
vis3 = vis1.merge(suara, how = 'right', right_on = 'Pilihan', left_on = 'Pilihan')
vis3 = vis3[['Harapan']]
vis3
```

```python
#vis3.plot(kind = 'barh', title = 'Jumlah Harapan Kosong 5 Top Calon Rektor')
vis3.plot(kind='barh',    # Plot a bar chart
        width=0.75,      # Mengatur lebar bar menjadi 75% dari space yang tersedia
        figsize=(6,4),  # Mengatur ukuran plot dalam inchi
        colormap='Set3', #Mengganti warna bar menjadi Biru Muda
        title = 'Jumlah Harapan Kosong 5 Top Calon Rektor') #Memberi judul
```

```python
# Visualisasi 4
# Membandingkan perolehan suara dari 6 Calon Rektor dari STEI
```

```python
#Urutan Suara Rektor
x = df.groupby('Pilihan').count().sort_values(by = ['Timestamp'],ascending = False)
x = x[['Timestamp']] #Variabel 5 calon rektor dgn suara terbanyak
x
```

```python
stei = x[3:4] + x[5:7] + x[12:13] + x[14:15] + x[17:18]
stei
```

```python
b = stei.merge(x, how = 'left',right_on = 'Pilihan', left_on = 'Pilihan')
b = b['Timestamp_y']
b
```

```python
vis4 = b.plot(kind='pie',autopct='%1.1f%%')
vis4.set_ylabel('')
vis4.set_xlabel('Persentase Perolehan Suara CaRek dari STEI')
vis4.plot(kind = 'pie')
```

```python
# Visualisasi 5
# Memvisualisasikan Persentase Suara Harapan Kosong dari 6 Ca-Rek STEI
```

```python
hstei = stei.merge(vis1, how = 'left',right_on = 'Pilihan', left_on = 'Pilihan')
hstei = hstei['Harapan']
hstei
```

```python
vis5 = hstei.plot(kind='pie',autopct='%1.1f%%')
vis5.set_ylabel('')
vis5.set_xlabel('Persentase Suara Harapan Kosong dari 6 Ca-Rek STEI')
vis5.plot(kind = 'pie')
```
