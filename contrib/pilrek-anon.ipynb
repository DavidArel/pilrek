```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df  = pd.read_csv('pilrek-anon.csv')
df.columns = ['Timestamp', 'Kategori', 'Umur', 'Carek', 'Alasan', 'Harapan']
```

```python
count_df = df['Carek'].value_counts()
new_df = df[df['Carek'].isin(count_df[count_df > count_df.mean()].index)]
new_df
```

```python
ax = new_df['Umur'].value_counts().plot(kind='pie',autopct='%1.1f%%')
ax.set_ylabel('')
ax.set_title('Persentase umur sample data')
```

```python
ax = new_df['Kategori'].value_counts().plot(kind='pie',autopct='%1.1f%%')
ax.set_ylabel('')
ax.set_title('Persentase kategori sample data')
```

```python
count = new_df['Alasan'].value_counts()
alasan = new_df[new_df['Alasan'].isin(count[count > 3].index)]
alasan.sample(5)
```

```python
chart = alasan['Alasan'].value_counts().plot(kind='bar')
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
chart.set_title('Alasan ')
```
